## Question 4

### UCR

````python
#Converting dates to Date Times
selected_ucr['New Dates'] = pd.to_datetime(selected_ucr["Date"]).dt.strftime("%Y%d%m")
selected_ucr.head()
````

````python
#Converting the New Dates to Strings
selected_ucr['New Dates'] = selected_ucr['New Dates'].astype(str)
````

````python
#Slicing the Dates to have the months
selected_ucr['MONTH'] = selected_ucr['New Dates'].str.slice(start = 6, stop = 8)
selected_ucr.head()
````

````python
#Finding Mean Precipation for Each UCR Month
precip_col = 'Precip (in)'
ucr_means = selected_ucr.groupby('MONTH')[precip_col].mean().reset_index()
ucr_means.head()
````

````python
#Finding Mean Air Temperature for Each UCR Month
air_temp_col = 'Avg Air Temp (F)'
ucr_means['Mean Air Temp (F)'] = 0
mean_air_temp = selected_ucr.groupby('MONTH')[air_temp_col].mean().reset_index()
ucr_means['Mean Air Temp (F)'] = mean_air_temp['Avg Air Temp (F)']
ucr_means.head()
````

````python
#Finding Mean Humidity for Each UCR Month
hum_col = 'Avg Rel Hum (%)'
ucr_means['Mean Rel Hum (%)'] = 0
mean_hum = selected_ucr.groupby('MONTH')[hum_col].mean().reset_index()
ucr_means['Mean Rel Hum (%)'] = mean_hum['Avg Rel Hum (%)']
ucr_means.head()
````

````python
#Renaming the Precip column in order to have a clear data set
new_names = {"Precip (in)": "Mean Precip (in)"}
ucr_means = ucr_means.rename(columns=new_names)
ucr_means.head()
````

## Question 6

````python
def fahr_to_celsius(temp_fahrenheit):
    """Function to convert Fahrenheit temperature into Celsius.

    Parameters
    ----------

    temp_fahrenheit: int | float
        Input temperature in Fahrenheit (should be a number)

    Returns
    -------

    Temperature in Celsius (float)
    """

    #Convert the Fahrenheit into Celsius
    converted_temp = (temp_fahrenheit - 32) / 1.8

    return converted_temp
````
